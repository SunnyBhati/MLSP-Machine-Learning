## Part-2

The pdf for a uniform distribution is as follows:
  $f_X(x) = 1/(b-a), \ b \leq x \leq a$ .
 In order to estimate the parameter a and b, we need to first formulate the MLE.

For a uniform distribution, the Maximum likelihood estimation is $L=\underset{a,b}{\operatorname{argmax}} (1/(b-a))^n$

Taking logarithm for the maximum likelohood estimation, we get $\log L= -n*\log(b-a)$

Differentiating w.r.t b, we get $\frac{\partial L}{\partial b}=\frac{-n}{b-a}$. Derivative w.r.t b is monotonically decreasing, therefore the smallest value of b will maximize the likelihood function which is $\max[92, 95.8, 91.3, 94.1, 90.9]$

Differentiating w.r.t a, we get $\frac{\partial L}{\partial a}=\frac{n}{b-a}$. Similarly, derivative w.r.t b is monotonically increasing function , therefore the largest value of a will maximise the likelihood function which is $\min[92, 95.8, 91.3, 94.1, 90.9]$

Therefore, the parameters a and b are 90.9 and 95.8 respectively

Let us consider the professor used a multiple of 10 to distribute A+ grades. Therefore, the new adjusted values for a and b would be 90 and 100 respectively. Since, a has to be $\min[92, 95.8, 91.3, 94.1, 90.9]$ and a multiple of 10 as per the assumption. Therefore, a is equal to 90.

b has to be $\max[92, 95.8, 91.3, 94.1, 90.9]$ and a multiple of 10 as per the assumption. Therefore, b is equal to 100.

The assumption of a multiple of 10 extends the interval to a broad interval from 90 to 100.

#### Comments for Question 2

In other words, high kurtosis value indicates the data is not uniformly distributed.Basically,high kurtosis value indicates low noise and low resolution. 

From the kurtosis values, it is to be noted that the luddy1 image has the highest kurtosis value implying sharper image compared to luddy2 image and luddy3 image. Luddy2 image with the lowest kurtosis value is the most shaken image and more Gaussian image compared to the luddy 1 and luddy 3.

In terms of sharpness, luddy1 image> luddy3 image > luddy2 image.

## Function for gradient descent and calculation of eigen vector

```python
import numpy as np
import math

def eigen_vector(random_samples):
    w_vector_1=np.random.rand(np.shape(random_samples)[0],1)
    w_vector_1=w_vector_1/math.sqrt(np.sum(np.square(w_vector_1)))

    z1=np.array(random_samples)
    z1_transpose=z1.transpose()
    learning_parameter=0.01
    a=0

    for i in range(10000):
        gh=np.matmul(z1,z1_transpose)
        gh1=np.matmul(gh,w_vector_1)*2 *learning_parameter
        wnew=[[w_vector_1[i][j] +gh1[i][j]  for j in range(len(w_vector_1[0]))] for i in range(len(w_vector_1))]
        wnew1=np.true_divide(wnew,math.sqrt(np.sum(np.square(wnew))))
        a=math.sqrt(np.sum(np.square(wnew1- w_vector_1)))
        if a < 0.000001:
            break
        else:
            w_vector_1 = wnew1
    return(w_vector_1)
```

## Computing the first eigen_vector with the largest eigenvalue

```python
firsteigenvector_finalmatrix=eigen_vector(Xmatrix_final)
print("The first eigenvector is:",firsteigenvector_finalmatrix)
```
